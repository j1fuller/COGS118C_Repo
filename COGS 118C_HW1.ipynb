```python
import numpy as np
import matplotlib.pyplot as plt

# Create time vector
t = np.linspace(0, 10, 1000)  # 1000 points between 0 and 10s
T = 2  # For function 5

# Create figure with subplots
plt.figure(figsize=(15, 25))

# a. f(t) = 3sin(3t) - 2
plt.subplot(7, 1, 1)
y1 = 3 * np.sin(3*t) - 2
plt.plot(t, y1)
plt.title('f(t) = 3sin(3t) - 2')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.grid(True)

# b. f(t) = (t-5)^2 + 2
plt.subplot(7, 1, 2)
y2 = (t - 5)**2 + 2
plt.plot(t, y2)
plt.title('f(t) = (t-5)^2 + 2')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.grid(True)

# c. f(t) = cos^3(2πt/√3)
plt.subplot(7, 1, 3)
y3 = np.cos(2*np.pi*t/np.sqrt(3))**3
plt.plot(t, y3)
plt.title('f(t) = cos^3(2πt/√3)')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.grid(True)

# d. f(t) = exp[-sin^2(3πt)]
plt.subplot(7, 1, 4)
y4 = np.exp(-np.sin(3*np.pi*t)**2)
plt.plot(t, y4)
plt.title('f(t) = exp[-sin^2(3πt)]')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.grid(True)

# e. f(t) = sin([t/T]+1.3) + cos(5t/T)
plt.subplot(7, 1, 5)
y5 = np.sin(t/T + 1.3) + np.cos(5*t/T)
plt.plot(t, y5)
plt.title('f(t) = sin([t/T]+1.3) + cos(5t/T), T=2')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.grid(True)

# f. f(t) = sin(1/t)
plt.subplot(7, 1, 6)
# Avoid division by zero by starting t slightly above 0
t_modified = np.linspace(0.1, 10, 1000)
y6 = np.sin(1/t_modified)
plt.plot(t_modified, y6)
plt.title('f(t) = sin(1/t)')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.grid(True)

# g. Sum series f(t) = Σ(1/n)*sin(nt), n=1 to N=100
plt.subplot(7, 1, 7)
y7 = np.zeros_like(t)
N = 100
for n in range(1, N+1):
    y7 += (1/n) * np.sin(n*t)
plt.plot(t, y7)
plt.title('f(t) = Σ(1/n)*sin(nt), N=100')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.grid(True)

plt.tight_layout()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define parameters
A = 0.005  # Amplitude in microvolts
f = 5      # Frequency in Hz
t = np.linspace(0, 1, 1000)  # Time vector from 0 to 1 second, 1000 points

# Calculate LFP signal
f1 = A * np.sin(2 * np.pi * f * t)

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(t, f1)

# Add labels and title
plt.xlabel('Time (s)')
plt.ylabel('LFP Amplitude (µV)')
plt.title('Single Neuron LFP Response')

# Add grid for better readability
plt.grid(True)

# Ensure tight layout
plt.tight_layout()
```
